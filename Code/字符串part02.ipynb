```python
# 151.反转字符串中的单词
class Solution:
    def single_reverse(self, s, start, end):
        while start < end:
            s[start], s[end] = s[end], s[start]
            start += 1
            end -= 1

    def reverseWords(self, s):
        result = ""
        fast = 0
        # 1. 首先将原字符串反转并且除掉空格, 并且加入到新的字符串当中
        # 由于Python字符串的不可变性，因此只能转换为列表进行处理
        s = list(s)
        s.reverse()
        while fast < len(s):
            if s[fast] != " ":
                if len(result) != 0:
                    result += " "
                while fast < len(s) and s[fast] != " ":
                    result += s[fast]
                    fast += 1
            else:
                fast += 1
        # 2.其次将每个单词进行翻转操作
        slow = 0
        fast = 0
        result = list(result)
        while fast <= len(result):
            if fast == len(result) or result[fast] == " ":
                self.single_reverse(result, slow, fast - 1)
                slow = fast + 1
                fast += 1
            else:
                fast += 1

        return "".join(result)
```

```python
# 55.右旋字符串
class Solution:
    def revserse(self, s, begin, end):
        left = begin
        right = end
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
        return s
    def rightPopeStr(self, s, k):
        s = list(s)
        length = len(s)
        s = self.revserse(s, 0, length - 1 - k)
        s = self.revserse(s, length - k, length - 1)
        s = self.revserse(s, 0, length - 1)
        return ''.join(s)
```

```python
# KMP算法流程
class Solution(object):
    def strStr(self, haystack, needle):
        """
        :type haystack: str
        :type needle: str
        :rtype: int
        """
        def kmpNext(s):
            j = -1
            i = 0
            s = list(s)
            next = [0] * len(s)
            next[i] = j
            for i in range(1, len(s)):
                while(j >= 0 and s[i] != s[j + 1]):
                    j = next[j]
                if s[i] == s[j + 1]:
                    j += 1
                next[i] = j
            return next

        next = kmpNext(needle)
        # print(next)
        j = -1
        if haystack == needle:
            return 0
        for i in range(len(haystack)):
            # print(i,j+1,j)
            while j >= 0 and haystack[i] != needle[j + 1]:
                j = next[j]
            if needle[j + 1] == haystack[i]:
                j += 1
            if j == len(needle) - 1:
                return i - len(needle) + 1
        return -1
        
```

```python
class Solution(object):
    def repeatedSubstringPattern(self, s):
        """
        :type s: str
        :rtype: bool
        """
        j = -1
        i = 0
        s = list(s)
        next = [0] * len(s)
        next[0] = j
        for i in range(1, len(s)):
            while j >= 0 and s[i] != s[j + 1]:
                j = next[j]
            if s[i] == s[j + 1]:
                j += 1
            next[i] = j

        strPre = next[-1]
        # print(strPre)
        if strPre == -1 or (strPre == 0 and len(s) != 2):
            return False
        elif len(s) == 1 or (len(s) % (len(s) - strPre - 1) == 0):
            return True
        else:
            return False 
        
```
