二分查找（循环不变量）

```python
# 704.二分查找、左闭右开
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left = 0
        right = len(nums)
        while left < right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid
        return -1
```

```python
# 704.二分查找、左闭右闭
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left = 0
        right = len(nums) - 1
        while left <= right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return -1
```

```python
# 35.搜索插入位置（不满足条件的位置就是插入的位置）
class Solution(object):
    def searchInsert(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        left = 0
        right = len(nums)
        while left < right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                left = mid + 1
            else:
                right = mid
        return right
```

```python
# 34.在排序数组中查找元素的第一个和最后一个位置,最终返回的值是哪个不太确定，重点题目
class Solution(object):
    def searchRange(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        def searchLeft(nums, target):
            left = 0
            right = len(nums)
            while left < right:
                mid = left + (right - left) // 2
                if nums[mid] < target:
                    left = mid + 1
                else:
                    right = mid
            return right

        def searchRight(nums, target):
            left = 0
            right = len(nums) - 1
            while left <= right:
                mid = left + (right - left) // 2
                if nums[mid] <= target:
                    left = mid + 1
                else:
                    right = mid - 1
            return right
        leftIndex = searchLeft(nums, target)
        rightIndex = searchRight(nums, target)
        if leftIndex > rightIndex:
            return [leftIndex, rightIndex]
        return [-1, -1]
          
        
```

```python
# 69.x的平方根
class Solution(object):
    def mySqrt(self, x):
        """
        :type x: int
        :rtype: int
        """
        left = 0
        right = x + 1
        while left < right:
            mid = left + (right - left) // 2
            if mid * mid == x or (mid * mid < x and (mid + 1) * (mid + 1) > x):
                return mid
            elif mid * mid < x:
                left = mid + 1
            else:
                right = mid
        return -1
```

```python
# 367.有效的完全平方数
class Solution(object):
    def isPerfectSquare(self, num):
        """
        :type num: int
        :rtype: bool
        """
        left = 0
        right = num + 1
        while left < right:
            mid = left + (right - left) // 2
            if mid * mid == num:
                return True
            elif mid * mid < num:
                left = mid + 1
            else:
                right = mid
        return False
```

移除元素（快慢指针）

```python
# 27.移除元素
class Solution(object):
    def removeElement(self, nums, val):
        """
        :type nums: List[int]
        :type val: int
        :rtype: int
        """
        fastIndex = 0
        slowIndex = 0
        while fastIndex < len(nums):
            if nums[fastIndex] == val:
                fastIndex += 1
            else:
                nums[slowIndex] = nums[fastIndex]
                slowIndex += 1
                fastIndex += 1
        return slowIndex
        
```

```python
# 26.删除排序数组中的重复项
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        slowIndex = 0
        fastIndex = 1
        while fastIndex < len(nums):
            if nums[slowIndex] == nums[fastIndex]:
                fastIndex += 1
            else:
                slowIndex += 1
                nums[slowIndex] = nums[fastIndex]
                fastIndex += 1
        return slowIndex + 1
```
