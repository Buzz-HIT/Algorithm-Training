```python
# 242.有效字母的异位词
class Solution(object):
    def isAnagram(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: bool
        """
        charDict = {}
        for i in s:
            if i not in charDict:
                charDict[i] = 1
            else:
                charDict[i] += 1
        for i in t:
            if i not in charDict:
                return False
            else:
                charDict[i] -= 1
                if charDict[i] == 0:
                    charDict.pop(i)
        if not charDict: # 判断字典是否为空，counts[c] = counts.get(c, 0) + 1
            return True
        return False
```

```python
#383.赎金信
class Solution(object):
    def canConstruct(self, ransomNote, magazine):
        """
        :type ransomNote: str
        :type magazine: str
        :rtype: bool
        """
        from collections import defaultdict
        hashmap = defaultdict(int)
        for i in magazine:
            hashmap[i] += 1
        for i in ransomNote:
            hashmap[i] -= 1
            if hashmap[i] < 0:
                return False
        return True


        
```

```python
# 49.字母异位词分组
class Solution(object):
    def groupAnagrams(self, strs):
        """
        :type strs: List[str]
        :rtype: List[List[str]]
        """
        from collections import defaultdict
        sorted_strs = []
        for s in strs:
            sorted_s = ''.join(sorted(s)) # 将字符串排序后重新拼接成字符串  
            sorted_strs.append(sorted_s)
        str_Dict = defaultdict(list)
        for i in range(len(strs)):
            str_Dict[sorted_strs[i]].append(strs[i])# 将原字符串加入对应的哈希表中
        solution = list(str_Dict.values())# 取出哈希表中的所有值并转换为列表
        return solution
        
```

```python
# 438.找到字符串中所有字母的异位词
class Solution(object):
    def findAnagrams(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: List[int]
        """
        from collections import defaultdict
        hashP = defaultdict(int)
        hashS = defaultdict(int)
        ans = []
        if len(s) < len(p):
            return []
        for i in p:
            hashP[i] += 1
        for i in range(len(p)):
            hashS[s[i]] += 1
        left = 0
        right = left + len(p) - 1
        while right < len(s):
            # print(hashP, hashS)
            if hashP == hashS:
                ans.append(left)
            right += 1
            if right < len(s):
                hashS[s[right]] += 1
                hashS[s[left]] -= 1
                if hashS[s[left]] == 0:
                    hashS.pop(s[left])
                left += 1
        return ans
                
```
