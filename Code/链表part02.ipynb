```python
# 24.两两交换链表中的节点
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def swapPairs(self, head):
        """
        :type head: Optional[ListNode]
        :rtype: Optional[ListNode]
        """
        if head == None:
            return
        dummyHead = ListNode(next=head)
        cur = dummyHead
        while cur.next and cur.next.next:
            index1 = cur.next
            index2 = cur.next.next
            cur.next = index2
            index1.next = index2.next
            index2.next = index1
            cur = index1
        return dummyHead.next
        
```

```python
# 19.删除链表的倒数第N个节点
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution(object):
    def removeNthFromEnd(self, head, n):
        """
        :type head: Optional[ListNode]
        :type n: int
        :rtype: Optional[ListNode]
        """
        if head == None:
            return
        dummyHead = ListNode(next=head)
        fastIndex = dummyHead
        slowIndex = dummyHead
        for i in range(0, n):
            fastIndex = fastIndex.next
        while fastIndex.next:
            slowIndex = slowIndex.next
            fastIndex = fastIndex.next
        slowIndex.next = slowIndex.next.next
        return dummyHead.next
```

```python
# 160.相交链表
# Definition for singly-linked list.
# class ListNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution(object):
    def getIntersectionNode(self, headA, headB):
        """
        :type head1, head1: ListNode
        :rtype: ListNode
        """
        def culLength(head):
            lenth = 0
            cur = head
            while cur:
                lenth += 1
                cur = cur.next
            return lenth
        lenthA = culLength(headA)
        lenthB = culLength(headB)
        if lenthA > lenthB:
            lenthA, lenthB = lenthB, lenthA
            headA, headB = headB, headA
        curA = headA
        curB = headB
        for i in range(0, lenthB - lenthA):
            curB = curB.next
        while curA and curB:
            if curA == curB:
                return curA
            curA = curA.next
            curB = curB.next
        return None

        
        
```
